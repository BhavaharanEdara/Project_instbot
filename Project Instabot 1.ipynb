```python
# IMPORT ALL THE REQURIED MODULES
from selenium import webdriver
from time import sleep
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import warnings

#IGNORING THE WARNINGS BY PYTHON
warnings.filterwarnings('ignore', category=DeprecationWarning)

# CREATING A CLASS INSTABOT WHICH CONTAINS MAIN METHODS OF THE INSTABOT
class instabot():
    def __init__(self,username,password):
        #CREATE THE CHROME DRIVER
        self.driver = webdriver.Chrome('C:\chro\chromedriver')
        self.login(username,password)
        
        
    #FUNCTION TO LOGIN TO ACCOUNT
    def login(self,username,password):
        self.driver.get('https://www.instagram.com/')
        
        #INPUT OF USERNAME
        WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.NAME, "username")))
        user_box = self.driver.find_element_by_name('username')
        user_box.send_keys(username)
        
        #INPUT OF PASSWORD
        pass_box = self.driver.find_element_by_name('password')
        pass_box.send_keys(password)
        self.driver.find_element_by_id("loginForm").click()
        
        
        # clcicking Not Now for save info and Turn on Notifications
        WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "cmbtv")))
        self.driver.find_element_by_class_name('cmbtv').click()
        WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "HoLwm")))
        self.driver.find_element_by_class_name('HoLwm').click()
    
    
    #SEARCHING FOR A GIVEN INPUT   
    def search(self,input):
        #WAIT UNTIL SEARCH BAR IS LOADED
        WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "DljaH")))
        search_box = self.driver.find_element_by_class_name('DljaH')
        
        #CLEAR THE SEARCH BOX IT MAY CONTAIN ANY OTHER INPUT
        search_box.clear()
        search_box.send_keys(input)
    
    #PRINTING THE HANDLES DISPLAIED AFTER SEARCHING FOR ELEMENT 
    def printing_handles(self):
        #WAIT UNTIL THE SEARCHBAR HANDLES LOADED
        WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "-qQT3")))
        handles = self.driver.find_elements_by_xpath("//div[contains(@class, '_7UhW9') and contains(@class, 'xLCgt') and contains(@class, 'qyrsm') and contains(@class, ' KV-D4 ') and contains(@class, 'uL8Hv')]")
        
        #PRINTING THE HANDLES
        for x in handles:
            a = x.get_attribute('innerHTML')
            if a != None:
                if a[0] == '#':
                    print(a[1:])
                else:
                    print(a)
                    
    #OPENING THE REQUIRED INSTAGRAM HANDLE 
    def opening_handle(self,inp):
        #USING SEARCH FUNCTION OF INSTABOT 
        self.search(inp)
        sleep(2)
        self.driver.find_element_by_class_name('-qQT3').click()
        
        
    #CLICKING ON FOLLOW
    def following_handle(self):
        sleep(2)
        follow_btn = self.driver.find_element_by_class_name('_6VtSN')
        if follow_btn.get_attribute('innerHTML') == 'Follow':
            follow_btn.click()
            sleep(2)
        else:
            print('you are already following this Instagram handle')
    
    
    #UNFOLLOWING THE PAGE
    def unfollowing_handle(self):
        sleep(2)
        unfollow_btn = self.driver.find_element_by_class_name('_6VtSN')
        if unfollow_btn.get_attribute('innerHTML') == 'Follow' :
            print('you have already unfollowed this Instagram handle')
        else:
            #OPENING UNFOLLOW POP UP
            unfollow_btn.click()
            WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "-Cab_")))
            #CLICKING ON UNFOLLOW
            self.driver.find_element_by_class_name("-Cab_").click()
    
    
    #LIKE THE TOP 'n' POSTS
    def like_n_posts(self,n):
        sleep(2)
        #OPENING THE FIRST POST
        firstpost = self.driver.find_element_by_class_name('v1Nh3')
        firstpost.click()
        
        #LOOP FOR LIKING THE TOP 30 POSTS
        for x in range(n):
            #WAIT UNTILL THE LIKE BUTTON IS LOADED
            WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.XPATH, "/html/body/div[6]/div[3]/div/article/div/div[2]/div/div/div[2]/section[1]/span[1]/button/div")))
            c = self.driver.find_elements_by_xpath('/html/body/div[6]/div[3]/div/article/div/div[2]/div/div/div[2]/section[1]/span[1]/button/div')
            
            if len(c) == 2:
                #LIKING THE POST
                WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "fr66n")))
                self.driver.find_element_by_class_name('fr66n').click()
                #CLICK ON THE '>' ARROW MARK FOR NEXT POST
                WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "l8mY4")))
                self.driver.find_element_by_class_name('l8mY4').click()
            elif len(c) == 1:
                print('already Liked post '+str(x+1))
                #CLICK ON THE '>' ARROW MARK FOR NEXT POST
                WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "l8mY4")))
                self.driver.find_element_by_class_name('l8mY4').click()
                
        #CLOSING THE POST WINDOW AFTER LIKING 30 POSTS
        self.driver.find_element_by_class_name('NOTWr').click()
        
        
    
    #UNLIKE THE TOP 'n' POSTS
    def unlike_n_posts(self,n):
        sleep(2)
        #OPENING THE FIRST POST
        firstpost = self.driver.find_element_by_class_name('v1Nh3')
        firstpost.click()
        
        #LOOP FOR UNLIKING THE TOP 30 POSTS
        for x in range(n):
            WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.XPATH, "/html/body/div[6]/div[3]/div/article/div/div[2]/div/div/div[2]/section[1]/span[1]/button/div")))
            c = self.driver.find_elements_by_xpath('/html/body/div[6]/div[3]/div/article/div/div[2]/div/div/div[2]/section[1]/span[1]/button/div')
            
            if len(c) == 1:
                self.driver.find_element_by_class_name('fr66n').click()
                #CLICK ON THE '>' ARROW MARK FOR NEXT POST
                WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "l8mY4")))
                self.driver.find_element_by_class_name('l8mY4').click()
            else:
                print('already unLiked post'+str(x+1))
                #CLICK ON THE '>' ARROW MARK FOR NEXT POST
                WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "l8mY4")))
                self.driver.find_element_by_class_name('l8mY4').click()
        
        #CLOSING THE POST WINDOW AFTER LIKING 30 POSTS
        self.driver.find_element_by_class_name('NOTWr').click()
    
    
    #PRINT THE FIRST 'n' FOLLOWERS
    def print_n_followers(self,n):
        #WAIT UNTILL THE FOLLOWERS BUTTTON IS LOADED
        WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="react-root"]/section/main/div/header/section/ul/li[2]')))
        sleep(3)
        
        #CLCKING ON FOLLOWERS BUTTON
        self.driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/header/section/ul/li[2]').click()
        WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, '_0imsa')))
        followers = self.driver.find_elements_by_class_name('_0imsa')
        
        # SCROLL DOWN TILL I GET 'n' FOLLOWERS LOADED
        while len(followers)<n:
            self.driver.execute_script("""var myElement = document.getElementsByClassName('isgrP');myElement[0].scrollBy(0,20000);""")
            sleep(2)
            followers = self.driver.find_elements_by_class_name('_0imsa')
        #PRINTING THE FOLLOWERS
        
        for x in range(500):
            print(x+1,'. ',followers[x].get_attribute('innerHTML'))
        self.driver.find_element_by_xpath('/html/body/div[6]/div/div/div[1]/div/div[2]/button').click()
    
    
    def get_my_followers(self):
        #CLICKING ON MY PROFILE PICTURE VISIBLE AT THE TOP RIGHT CORNER
        self.driver.find_element_by_xpath('//*[@id="react-root"]/section/nav/div[2]/div/div/div[3]/div/div[6]/span/img').click()
        WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="react-root"]/section/nav/div[2]/div/div/div[3]/div/div[6]/div[2]/div[2]/div[2]/a[1]/div')))
        
        #CLICKING ON MY PROFILE
        myinstabot.driver.find_element_by_xpath('//*[@id="react-root"]/section/nav/div[2]/div/div/div[3]/div/div[6]/div[2]/div[2]/div[2]/a[1]/div').click()
        
        sleep(3)
        #WAIT UNTILL FOLLOWERS BUTTON IS LOADED
        WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="react-root"]/section/main/div/header/section/ul/li[2]/a/span')))
        
        #GETTING THE HTML OF THE NO_OF FOLLOWERS
        html = (self.driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/header/section/ul/li[2]/a/span').get_attribute('outerHTML'))
        
        #OBTAINING THE STRING OF NO OF FOLLOWERS
        no_of_followers_string = BeautifulSoup(html,'html.parser').span['title']
        #REMOVING ALL THE COMMAS IN THE NO OF FOLLOWERS
        no_of_followers_modifing = ''
        for x in no_of_followers_string:
            if x != ',':
                no_of_followers_modifing += x
        
        #OPENING THE FOLLOWERSS POPUP WINDOW
        self.driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/header/section/ul/li[2]/a').click()
        WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'isgrP')))

        followers_pop = self.driver.find_element_by_class_name('isgrP')
        WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, '_0imsa')))
        
        #SCROLLING UNTILL I GET ALL MY FOLLOWERS LOADED
        followers = self.driver.find_elements_by_class_name('_0imsa')
        while len(followers)!=int(no_of_followers_modifing):
            self.driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight",followers_pop)
            sleep(1)
            followers = self.driver.find_elements_by_class_name('_0imsa')
            
        # STORING THE FOLLOWERS IN A DICTIONARY AND RETURNING IT
        dict_of_followers = {}
        for x in followers:
            if x.get_attribute('innerHTML') not in dict_of_followers:
                dict_of_followers[x.get_attribute('innerHTML')] = 1
        #CLOSING THE FOLLOWERS POPUP WINDOW
        self.driver.find_element_by_xpath('/html/body/div[6]/div/div/div[1]/div/div[2]/button').click()
        return dict_of_followers
        
       
    def get_im_following(self):
        #CLICKING ON MY PROFILE PICTURE VISIBLE AT THE TOP RIGHT CORNER
        self.driver.find_element_by_xpath('//*[@id="react-root"]/section/nav/div[2]/div/div/div[3]/div/div[6]/span/img').click()
        WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="react-root"]/section/nav/div[2]/div/div/div[3]/div/div[6]/div[2]/div[2]/div[2]/a[1]/div')))
        
        #CLICKING ON MY PROFILE
        myinstabot.driver.find_element_by_xpath('//*[@id="react-root"]/section/nav/div[2]/div/div/div[3]/div/div[6]/div[2]/div[2]/div[2]/a[1]/div').click()
        
        #WAIT UNTILL FOLLOWING BUTTON IS LOADED
        sleep(3)
        WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="react-root"]/section/main/div/header/section/ul/li[3]/a/span')))

        #OBTAINING THE NO OF FOLLOWEING AND REMOVING THE COMMMAS FROM IT
        no_of_following_string = myinstabot.driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/header/section/ul/li[3]/a/span').get_attribute('innerHTML')
        no_of_following_modifing = ''
        for x in no_of_following_string:
            if x != ',':
                no_of_following_modifing += x
                
        #OPEING FOLLOWING POP UP
        self.driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/header/section/ul/li[3]/a').click()
        
        #WAIT UNTIL THE FOLLOWING POPUP IS LOADED
        WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'isgrP')))
        following_pop = self.driver.find_element_by_class_name('isgrP')
        
        #WAIT UNTILL THE FOLLOWING USER NAMES LOADED
        WebDriverWait(self.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, '_0imsa')))
        following = self.driver.find_elements_by_class_name('_0imsa')
        
        #SCROLLING UNTILL I GET ALL USERNAMES IM FOLLOWING LOADED
        while len(following)!=int(no_of_following_modifing):
            self.driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight",following_pop)
            sleep(1)
            following = self.driver.find_elements_by_class_name('_0imsa')
        #APPEND ALL THE USERNAMES INTO A LIST AND RETURNING
        list_of_following = []
        for x in following:
            a = x.get_attribute('innerHTML')
            list_of_following.append(a)
        #CLOSING THE FOLLOWING POPUP WINDOW
        self.driver.find_element_by_xpath('/html/body/div[6]/div/div/div[1]/div/div[2]/button').click()
        return list_of_following

    
# LOGING IN BY CREATING INSTABOT OBJECT
#PLACE YOUR USERNAME, PASSWORD
myinstabot = instabot('USERNAME','PASSWORD')
sleep(2)

#SEARCHING FOR FOOD
myinstabot.search('food')


#PRINTING HANDLES OF FOOD
print('Handles of "food" searched in the search bar')
myinstabot.printing_handles()
print('')


#SEARCHING AND OPENING 'So Delhi'
myinstabot.opening_handle('So Delhi')
sleep(2)


#FOLLOWING 'So Delhi'
myinstabot.following_handle()

#UNFOLLOWING 'So Delhi'
myinstabot.unfollowing_handle()

#SEARCHING AND OPENING 'dilsefoodie'
myinstabot.opening_handle('dilsefoodie')


#LIKING TOP 30 POSTS
myinstabot.like_n_posts(30)

#UNLIKING TOP 30 POSTS
myinstabot.unlike_n_posts(30)

#PRINTING THE TOP 500 FOLLOWERS OF 'foodtalkindia'
myinstabot.opening_handle('foodtalkindia')
print('First 500 followers of foodtalkinda')
myinstabot.print_n_followers(500)
print('')

#PRINTING THE TOP 500 FOLLOWERS OF 'So Delhi'
myinstabot.opening_handle('So Delhi')
print('First 500 followers of So Delhi')
myinstabot.print_n_followers(500)
print('')

#PEOPLE WHOME I AM FOLLOWING BUT NOT FOLLOWING ME
my_followers_dict = myinstabot.get_my_followers()
im_following_list = myinstabot.get_im_following()
# THE REQUIRED USERNAMES IS STORED IN A DICTONARY
required_dict = {}
for x in im_following_list:
    if x not in my_followers_dict:
        required_dict[x] = 1

        
#OPENONG FOODTALKINDIA AND GETTING ALL THE FOLLLOWERS
myinstabot.opening_handle('foodtalkindia')

#WAT UNTIL THE FOLLOWERS BUTTON IS LOADED
WebDriverWait(myinstabot.driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="react-root"]/section/main/div/header/section/ul/li[2]/a/span')))
html_of_followers = myinstabot.driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/header/section/ul/li[2]/a/span').get_attribute('outerHTML')
#USING BEAUTIFULSOUP FINDING THE NO OF FOLLOWERS IN STRING 
parsedhtml = BeautifulSoup(html_of_followers,'html.parser')
no_of_followers_string = parsedhtml.span['title']

#MODIFING THE NO OF FOLLOWERS BY REMOVING COMMAS
no_of_followers_modified = ''
for x in no_of_followers_string:
    if x!=',':
        no_of_followers_modified += x
        
#OPENING FOLLOWERS POPUP
myinstabot.driver.find_element_by_xpath('//*[@id="react-root"]/section/main/div/header/section/ul/li[2]/a').click()

#wait until the popup window is loaded
WebDriverWait(myinstabot.driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'isgrP')))
followers_pop = myinstabot.driver.find_element_by_class_name('isgrP')

#WAIT UNTILL THE USERNAMES ARE LOADED
followers = myinstabot.driver.find_elements_by_class_name('_0imsa')

#SCROLL UNTILL ALL THE FOLLOWERS ARE LOADED
while len(followers) <= int(no_of_followers_modified):
    myinstabot.driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight",followers_pop)
    followers = myinstabot.driver.find_elements_by_class_name('_0imsa')
    
    #USING 'required_dict' WE SEPERATE THE PEOPLE WHOME I AM FOLLOWING BUT NOT FOLLOWING ME
    for x in followers:
        if x.get_attribute('innerHTML') in required_dict:
            print(x.get_attribute('innerHTML'))
        
#CLOSIG THE FOLLOWERS POP UP
myinstabot.find_element_by_xpath('/html/body/div[6]/div/div/div[1]/div/div[2]/button').click()


#GETTING STORY OF CODING.NINJAS
myinstabot.opening_handle('coding.ninjas')
WebDriverWait(myinstabot.driver, 10).until(EC.presence_of_element_located((By.CLAAS_NAME, 'CfWVH')))
try:
    if int(driver.find_element_by_xpath('//div[contains(@class, "h5uC0")]/canvas').get_attribute('height'))==210:
        print('You have not seen the story yet! The story will be shown to you now. check out the driver window')
        driver.find_element_by_xpath('//div[contains(@class, "h5uC0")]').click()
    elif int(driver.find_element_by_xpath('//div[contains(@class, "h5uC0")]/canvas').get_attribute('height'))==166:
        print('You have already seen the story!')
except NoSuchElementException:
    print('The user has no story!')
```
